```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
para={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'splitter':['best','random'],'random_state':[20,40,60]}
regressor=GridSearchCV(DecisionTreeRegressor(),para,refit=True,verbose=3,n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
result=regressor.cv_results_
table=pd.DataFrame.from_dict(result)
table
```

```python
table.to_excel('DTR_DataFrame.xlsx',index=False)
```

```python
print("The best scored parameter is: {}".format(regressor.best_params_))
```

```python
import pickle
pickle.dump(regressor,open("Grid_DTreeR.sav",'wb'))
```
