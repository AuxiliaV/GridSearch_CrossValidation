```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param={'kernel':['linear','poly','rbf','sigmoid'],'C':[10,100,1000,2000,3000,100000,200000,300000]}
regressor=GridSearchCV(SVR(), param, refit=True, verbose=3, n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
result=regressor.cv_results_
print(result)
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
import pickle
filename="SVM_Grid_Dataframe.csv"
pickle.dump(table,open(filename,'wb'))
```

```python
print("The best scored parameter {}".format(regressor.best_params_))
```

```python
pickle.dump(regressor,open("Grid_SVMR.sav",'wb'))
```

```python
model=pickle.load(open("Grid_SVMR.sav",'rb'))
out=model.predict([[35,40,2,1,1]])
out
```
